```python
import json
import pandas as pd

tmp = {}
arr = []

with open('C:\\xampp\\htdocs\\leaflet\\custom.geo.json', 'r') as f:
    data = json.load(f)

for i in data['features']:
    tmp['type'] = i['type']
    tmp['properties'] = {
        'name': i['properties']['NAME'],
        'fullname': i['properties']['NAME_LONG'],
        'code': i['properties']['ISO_A3']
    }
    tmp['geometry'] = i['geometry']
    arr.append(tmp)
    tmp = {}

final = {'features': arr}
arr = []

for i in final['features']:
    arr.append(i['properties']['fullname'])

data_gtdb = pd.read_excel('gtdb.xlsx')
countries = list(data_gtdb.loc[:,'country_txt'])
for i in arr:
    if i not in countries:
        print i

# with open('countries.geo.json', 'w') as outf:
#     json.dump(final, outf)
```
